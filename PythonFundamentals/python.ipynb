```python
student_ids = ('S101', 'S102', 'S103', 'S104') 
students = {
'S101': {'name': 'Asha', 'assignment': 78, 'test': 80, 'attendance': 92, 'hours': 8}, 
'S102': {'name': 'Ravi', 'assignment': 65, 'test': 68, 'attendance': 85, 'hours': 5},
'S103': {'name': 'Meena', 'assignment': 88, 'test': 90, 'attendance': 96, 'hours': 10},
'S104': {'name': 'Kiran', 'assignment': 55, 'test': 58, 'attendance': 78, 'hours': 4}
}
```

```python
def calculate_average(id):
    average = students[id]['assignment'] * 0.4 + students[id]['test'] * 0.6
    return average
```

```python
def calculate_risk(id):
    hours = students[id]['hours']
    attendance = students[id]['attendance']
    average = students[id]['average']
    if hours < 6 or attendance < 80 or average<60:
        risk = 'High'
    elif 6 <= hours < 9 or 80 <= attendance < 90 or 60 <= average < 75:
        risk = 'Medium'
    else:
        risk = 'Low'
    return risk
```

```python
for id in student_ids:
    students[id]['average'] = calculate_average(id)
    students[id]['risk'] = calculate_risk(id) 
```

```python
for id in student_ids:
    print(f"""
Student ID: {id},\nName: {students[id]['name']},
Average: {students[id]['average']},\nRisk: {students[id]['risk']}
Assignment: {students[id]['assignment']},\nTest: {students[id]['test']}
Attendance: {students[id]['attendance']},\nHours: {students[id]['hours']}""")
```
